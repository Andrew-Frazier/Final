```python
def comp(ite):
    avgs = []
    for i in range(df.shape[0]):
        if df.iloc[i,0].split("-")[1] == ite:
            avgs.append(df.iloc[i, 1])


    avgs.sort()
    val_splits = []
    total_splits = int(len(avgs) / 5) - 1
    print(len(avgs))
    print(total_splits)
    b = total_splits
    for _ in range(0, 4): # five possible splits
        val_splits.append(avgs[total_splits])
        total_splits += b

    print(val_splits)
    return val_splits

res = {}

for i in range(1, 10):
    res[i] = comp(str(i))
```
